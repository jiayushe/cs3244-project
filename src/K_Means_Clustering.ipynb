```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import preprocessing, neighbors
from sklearn.preprocessing import MinMaxScaler 
```

```python
df = pd.read_csv("../data/Mental_Health_Clean.csv")
scaler = MinMaxScaler()
scaler.fit(df[['How many employees does your company or organization have?']])
df['How many employees does your company or organization have?'] = scaler.transform(df[['How many employees does your company or organization have?']])
scaler.fit(df[['What is your age?']])
df['What is your age?'] = scaler.transform(df[['What is your age?']])
df.head()
```

```python
X = np.array(df.drop(['Have you been diagnosed with a mental health condition by a medical professional?'], 1))
y = np.array(df['Have you been diagnosed with a mental health condition by a medical professional?'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
clf = neighbors.KNeighborsClassifier(n_neighbors=11, weights = 'distance')
clf.fit(X_train, y_train)
```

```python
accuracy = clf.score(X_test, y_test)
accuracy
```
